```python
####Decision Tree Splitting --  Condition : X variables are Continuous

import pandas as pd
from tkinter import filedialog

# Split of data on attribute value
def split_data(attr,cond,node):
    left_node,right_node = pd.DataFrame(),pd.DataFrame()
    for i in range(0,len(node)):
        if(node.iloc[i][attr] < cond):
            left_node = left_node.append(node.iloc[i],ignore_index=True)
        else:
            right_node = right_node.append(node.iloc[i],ignore_index=True)
    return right_node,left_node

#Evaluation Coefficient for the split - Weighted sum of sqs of probs
def evaluation_coeff(node_data):
    eval_coeff = 0.0
    no_of_records =float(sum([len(data) for data in node_data]))
    for node in node_data:
        value = 0.0
        no_of_nrows = float (len(node))
        if(no_of_nrows > 0):
            for cls in cl:
                s = [data[pred] for i,data in node.iterrows()].count(cls)
                p  = s / no_of_nrows
                value = value + p * p
        eval_coeff = eval_coeff+ value * (no_of_nrows/no_of_records)
    return eval_coeff

#Determination of the split and best Evaluation coefficient:
def split_deter(node):
    eval = 0
    attr = None
    eval_df = pd.DataFrame()
    for list in node.columns:
        if (list!= pred):
            for i,row in node.iterrows():
                split = split_data(list,row[list],node)
                eval_split= evaluation_coeff(split)
                eval_df = eval_df.append(pd.Series([list,row[list],eval_split]),ignore_index=True)
                #print(list,row[list],eval_split)
                if(eval_split) > eval:
                    attr,val,eval= list,row[list],eval_split
    return attr,val,eval,eval_df



#Node data
#data = pd.ExcelFile(r"C:\Users\sugan\Desktop\Tree_data.xlsx",Header= True)
#node = data.parse("Tree",header=0)

data = pd.ExcelFile(filedialog.askopenfilename())
node = data.parse(data.sheet_names[0],header=0)
pred = node.columns[-1]

#Determining Classes in output variable
for i in range(0,len(node)):
    if(i==0):
        cl = list()
    else:
        cl.append(node.iloc[i][pred])
cl = set(cl)

#Getting the best Split
attr,val,split,eval_df1 = split_deter(node)

#Getting the midpoint of the split attribute Selected value
df1 = list(node[attr])
df1.sort()
ind = df1.index(val)
#Should not be final index
if (ind < df1.index(df1[-1])):
    a = list(set(eval_df1[eval_df1[1]==df1[ind-1]][2]))
    b = list(set(eval_df1[eval_df1[1]==df1[ind+1]][2]))
    if (float(a[0]) > float(b[0])):
        split_value = val - ((val - df1[ind-1])/2)
    else:
        split_value = val - ((val - df1[ind+1])/2)

#Dataframes with Eval.Coefficient values
print(eval_df1)
        
#Best Eval Coefficient:
print("\nBest attribute and Evaluation coefficent value")
print(attr,"<",val,"Eval-Coefficient",split)

#Best Splitvalue:
print("\n Best Split value")
print(attr,"<=", split_value)
```
